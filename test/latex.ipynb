## Latex examples

Select examples from https://en.wikibooks.org/wiki/LaTeX/Mathematics

Note latex2mathml package doesn't support everything.

```python
import ziamath as zm
```

```python
zm.Math.fromlatex(r'\frac{1 + \alpha}{1 + \beta}')
```

```python
zm.Math.fromlatex(r'k_{n+1} = n^2 + k_n^2 - k_{n-1}')
```

```python
zm.Math.fromlatex('f(n) = n^5 + 4n^2 + 2 |_{n=17}')
```

```python
zm.Math.fromlatex(r'\frac{n!}{k!(n-k)!} = \binom{n}{k}')
```

```python
zm.Math.fromlatex(r'\frac{\frac{1}{x}+\frac{1}{y}}{y-z}')
```

```python
zm.Math.fromlatex(r'\sqrt{\frac{a}{b}}')
```

```python
zm.Math.fromlatex(r'\sum_{i=1}^{10} t_i')
```

```python
zm.Math.fromlatex(r'\left(\frac{x^2}{y^3}\right)')
```

```python
zm.Math.fromlatex(r'\frac{\mathrm d}{\mathrm d x} \left( k g(x) \right)')
```

```python
zm.Math.fromlatex(r'''\begin{matrix}
  a & b & c \\
  d & e & f \\
  g & h & i
 \end{matrix}''')
```
